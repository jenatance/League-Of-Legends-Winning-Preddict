```python
import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import warnings
```

```python
diamond = pd.read_csv('../data/lol_ranked_10min.csv')
```

```python
diamond
```

```python
diamond['blueDeathsCopy'] = diamond['blueDeaths'] 
diamond['blueDeathsCopy'] = diamond.loc[:, 'blueDeaths']
```

```python
diamond['redDeathsCopy'] = diamond['redDeaths'] 
diamond['redDeathsCopy'] = diamond.loc[:, 'redDeaths']
```

```python
diamond
```

```python
print(diamond['blueDeathsCopy'].unique())
print(diamond['redDeathsCopy'].unique())
```

```python
diamond.blueDeathsCopy = diamond.blueDeathsCopy.replace(0, 1)
diamond.redDeathsCopy = diamond.redDeathsCopy.replace(0, 1)
```

```python
print(diamond['blueDeathsCopy'].unique())
print(diamond['redDeathsCopy'].unique())
```

```python
diamond["blueKDA"] = np.nan
diamond["redKDA"] = np.nan
```

```python
diamond.info()
```

```python
diamond['blueKDA'] = (diamond['blueKills'] + diamond['blueAssists']) / diamond['blueDeathsCopy']
```

```python
diamond['redKDA'] = (diamond['redKills'] + diamond['redAssists']) / diamond['redDeathsCopy']
```

```python
kda = diamond[['blueKDA','blueKills','blueDeaths','blueAssists','redKDA','redKills','redDeaths','redAssists']]
```
